```python
import pandas as pd
import numpy
import datetime
```

```python
xls = pd.ExcelFile('Manifest data file.xlsx')
df1 = pd.read_excel(xls, 'Stops')
df1.head(5)
```

```python
df = df1['Sch Date'].values
arr = []
for i in range(len(df)):
    a=df[i]
    try:
        val = str(int(a))
        date_obj = datetime.datetime.strptime(val ,'%Y%m%d')
        val = str(date_obj.date())
        arr.append(val)
    except:
        arr.append("NA")
print(type(arr))
df1['epoch Sch Date'] = arr
df1
```

```python
df = df1['Sch Time'].values
arr = []
for i in range(len(df)):
    a=df[i]
    try:
        val = int(a)
        h = str(int(val/60)).zfill(2)
        m = str(val%60).zfill(2)
        s = "00"
        val = h+":"+m+":"+s
        arr.append(val)
    except:
        arr.append("NA")
print(type(arr))
df1['epoch Sch Time'] = arr
df1
```

```python
df = df1[['epoch Sch Date','epoch Sch Time']].values
arr = []
for i in range(len(df)):
    a=df[i]
    try:
        val = str(a)
        val1 = val[2:12]
        val2 = val[15:23]
        val=str(val1+" "+val2+",72")
        dt_obj = datetime.datetime.strptime(val, '%Y-%m-%d %H:%M:%S,%f')  
        epoch_msec = dt_obj.timestamp()*1000
        arr.append(epoch_msec)
    except:
        arr.append("NA")
df1['Epoch Time'] = arr
df1
```
